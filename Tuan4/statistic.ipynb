## Câu 1: 
- a.	Lập bảng tần số?
- b.	Vẽ biểu đồ cột biểu diễn sự phân bố tần suất?
- c.	Tính số trung bình, độ lệch chuẩn, số trung vị và mốt?


```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu năng suất (tạ/ha)
nang_suat = [30, 30, 20, 25, 25, 45, 40, 40, 35, 50, 45, 35, 25, 45, 30, 30, 30, 40, 30, 25, 35, 35, 45, 35, 35, 35, 30, 40, 40, 40, 35]

# a. Lập bảng tần số
tanso = {x: nang_suat.count(x) for x in nang_suat}
print("Bảng tần số:")
print("Năng suất (tạ/ha) | Số tỉnh")
for key, value in sorted(tanso.items()):
    print("{:<19} | {}".format(key, value))

# b. Vẽ biểu đồ cột
plt.hist(nang_suat, bins=range(20, 60, 5), edgecolor='black', alpha=0.7)
plt.xlabel('Năng suất (tạ/ha)')
plt.ylabel('Tần suất')
plt.title('Biểu đồ phân bố tần suất năng suất lúa')
plt.xticks(range(20, 60, 5))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# c. Tính số trung bình, độ lệch chuẩn, số trung vị và mốt
mean = np.mean(nang_suat)
std_dev = np.std(nang_suat)
median = np.median(nang_suat)
mode = max(set(nang_suat), key=nang_suat.count)

print("\nSố trung bình: {:.2f}".format(mean))
print("Độ lệch chuẩn: {:.2f}".format(std_dev))
print("Số trung vị: {:.2f}".format(median))
print("Mốt: {}".format(mode))
```

### Mẫu số liệu bệnh nhân đau mắt hột
- 21	17	22	18	20	17	15	13	15	20

### a.	Lập bảng tần số?
### b.	Vẽ biểu đồ cột biểu diễn sự phân bố tần suất?
### c.	Tính số trung bình, độ lệch chuẩn, số trung vị, tứ phân vị và mốt?



```python
import statistics
import matplotlib.pyplot as plt

# Dữ liệu tuổi của bệnh nhân đau mắt hột
tuoi = [21, 17, 22, 18, 20, 17, 15, 13, 15, 20,
        15, 12, 18, 17, 25, 17, 21, 15, 12, 18,
        16, 23, 14, 18, 19, 13, 16, 19, 18, 17]

# a. Lập bảng tần số
tanso = {x: tuoi.count(x) for x in tuoi}
print("Bảng tần số:")
print("Tuổi | Số lần xuất hiện")
for key, value in sorted(tanso.items()):
    print("{:<5} | {}".format(key, value))

# b. Vẽ biểu đồ cột
plt.hist(tuoi, bins=range(min(tuoi), max(tuoi) + 1), edgecolor='black', alpha=0.7)
plt.xlabel('Tuổi')
plt.ylabel('Tần suất')
plt.title('Biểu đồ phân bố tần suất tuổi bệnh nhân đau mắt hột')
plt.xticks(range(min(tuoi), max(tuoi) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# c. Tính số trung bình, độ lệch chuẩn, số trung vị, tứ phân vị và mốt
mean = statistics.mean(tuoi)
std_dev = statistics.stdev(tuoi)
median = statistics.median(tuoi)
quartiles = statistics.quantiles(tuoi)
mode = statistics.mode(tuoi)

print("\nSố trung bình: {:.2f}".format(mean))
print("Độ lệch chuẩn: {:.2f}".format(std_dev))
print("Số trung vị: {:.2f}".format(median))
print("Tứ phân vị: {}".format(quartiles))
print("Mốt: {}".format(mode))
```

### Câu 7
### Một cửa hàng bán xe ô tô thay đổi chiến lược kinh doanh vào cuối năm 2019. Số xe cửa hàng bán được mỗi tháng trong năm 2019 và 2020 được ghi lại ở bảng sau:
### Tháng	Năm 2019	Năm 2020
-  1	54	45
-  2	22	28
-  3	24	31
-  4	30	34
-  5	35	32
-  6	40	35
-  7	31	37
-  8	29	33
-  9	29	33
- 10	37	35
- 11	40	34
- 12	31	37
### a.	Tính số trung bình, độ lệch chuẩn, số trung vị, tứ phân vị và mốt của số lượng xe bán được trong năm 2019 và năm 2020?
### b.	Vẽ biểu đồ cột đôi biểu diễn số xe cửa hàng bán được mỗi tháng trong năm 2019 và 2020?
## c.	Nêu nhận xét về tác động của chiến lược kinh doanh mới lên số lượng xe bán ra hàng tháng?


```python
import pandas as pd
import numpy as np
import statistics

# Dữ liệu số lượng xe bán được trong năm 2019 và 2020
data = {
    'Tháng': np.arange(1, 13),
    'Năm 2019': [54, 22, 24, 30, 35, 40, 31, 29, 29, 37, 40, 31],
    'Năm 2020': [45, 28, 31, 34, 32, 35, 37, 33, 33, 35, 34, 37]
}

# Chuyển đổi dữ liệu thành DataFrame
df = pd.DataFrame(data)

# a. Thống kê số trung bình, độ lệch chuẩn, số trung vị, tứ phân vị và mốt
for year in ['Năm 2019', 'Năm 2020']:
    print(f"\n{year}:")
    print("Số trung bình:", df[year].mean())
    print("Độ lệch chuẩn:", df[year].std())
    print("Số trung vị:", df[year].median())
    print("Tứ phân vị:", np.percentile(df[year], [25, 50, 75]))
    print("Mốt:", statistics.mode(df[year]))


# c. Nhận xét về tác động của chiến lược kinh doanh mới
print("\nNhận xét:")
print("Tổng số xe bán được trong năm 2019:", df['Năm 2019'].sum())
print("Tổng số xe bán được trong năm 2020:", df['Năm 2020'].sum())
```

## Câu 20: 
### a.	Tính hệ số tương quan giữa nhiệt độ và khối lượng đường bị biến đổi? Nêu nhận xét về mối tương quan giữa hai đại lượng này?
### b.	Ước lượng đường hồi quy tuyến tính? Vẽ các điểm dữ liệu và đường hồi quy trên cùng một đồ thị? Ước lượng khối lượng trung bình của đường bị biến đổi được tạo ra khi nhiệt độ được mã hóa là 1.75OC?


```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
nhiet_do = np.array([1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0])
khoi_luong_duong = np.array([8.1, 7.8, 8.5, 9.8, 9.5, 8.9, 8.6, 10.2, 9.3, 9.2, 10.5])

# a. Tính hệ số tương quan
corr_coef = np.corrcoef(nhiet_do, khoi_luong_duong)[0, 1]
print("Hệ số tương quan giữa nhiệt độ và khối lượng đường:", corr_coef)
print("Nhận xét: Mối tương quan giữa nhiệt độ và khối lượng đường là", "tương đối mạnh" if abs(corr_coef) >= 0.7 else "yếu")

# b. Ước lượng đường hồi quy tuyến tính
slope, intercept = np.polyfit(nhiet_do, khoi_luong_duong, 1)
print("Đường hồi quy tuyến tính: y =", slope, "x +", intercept)

# Vẽ đồ thị
plt.scatter(nhiet_do, khoi_luong_duong, color='blue', label='Dữ liệu')
plt.plot(nhiet_do, slope * nhiet_do + intercept, color='red', label='Đường hồi quy')
plt.xlabel('Nhiệt độ (OC)')
plt.ylabel('Khối lượng đường (mg)')
plt.title('Biểu đồ dữ liệu và đường hồi quy tuyến tính')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# c. Ước lượng khối lượng trung bình khi nhiệt độ là 1.75OC
khoi_luong_trung_binh = slope * 1.75 + intercept
print("Ước lượng khối lượng trung bình của đường khi nhiệt độ là 1.75OC:", khoi_luong_trung_binh)
```
