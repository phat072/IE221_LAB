```python
import requests
from bs4 import BeautifulSoup

# Send a GET request
response = requests.get('https://openaccess.thecvf.com/WACV2024')

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the <dt class="ptitle"> tags
paper_titles_tags = soup.find_all('dt', class_='ptitle')

# Extract the text from these tags
paper_titles = [tag.text.strip() for tag in paper_titles_tags]

# Print the paper titles
for title in paper_titles:
    print(title)
```

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the stopwords from NLTK
nltk.download('punkt')
nltk.download('stopwords')

# Define a list of unimportant words in the field of technology
unimportant_words = ['the', 'a', 'an', 'in', 'is', 'of', 'and', 'for']

# Get the list of stopwords in English
stop_words = set(stopwords.words('english'))

# Add the unimportant words to the list of stopwords
stop_words.update(unimportant_words)

# Process each title
processed_titles = []
for title in paper_titles:
    # Tokenize the title into words
    words = word_tokenize(title)
    
    # Convert each word to lowercase and remove punctuation
    words = [word.lower() for word in words if word.isalpha()]
    
    # Remove stopwords and unimportant words
    words = [word for word in words if word not in stop_words]
    
    # Add the processed title to the list
    processed_titles.append(words)

# Print the processed titles
for title in processed_titles:
    print(title)
```

```python
import collections

# Flatten the list of processed titles into a single list of keywords
keywords = [word for title in processed_titles for word in title]

# Count the frequency of each keyword
keyword_freq = collections.Counter(keywords)

# Print the frequency of each keyword
for keyword, freq in keyword_freq.items():
    print(f'{keyword}: {freq}')
```

```python
import matplotlib.pyplot as plt

# Get the most common keywords
common_keywords = keyword_freq.most_common()

# Separate the keywords and their frequencies
keywords, frequencies = zip(*common_keywords)

# Plot a bar graph
plt.figure(figsize=(10, 5))
plt.bar(keywords, frequencies)
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.title('Frequency of Keywords')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the most common keywords
common_keywords = keyword_freq.most_common(5)  # Get only the top 5

# Separate the keywords and their frequencies
keywords, frequencies = zip(*common_keywords)

# Plot a bar graph
plt.figure(figsize=(10, 5))
plt.bar(keywords, frequencies)
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.title('Frequency of Top 5 Keywords')
plt.xticks(rotation=90)
plt.show()

# Display the top 5 keywords
print('Top 5 keywords:')
for keyword, freq in common_keywords:
    print(f'{keyword}: {freq}')
```

```python
import requests
from bs4 import BeautifulSoup
import collections
import matplotlib.pyplot as plt

# Send a GET request
response = requests.get('https://openaccess.thecvf.com/WACV2024')

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the <form> tags with class 'authsearch'
author_tags = soup.find_all('form', class_='authsearch')

# Extract the author names from these tags
author_names = [tag.input['value'] for tag in author_tags]

for author in author_names:
    print(author)
```

```python
# Count the frequency of each author name
author_freq = collections.Counter(author_names)

for author, freq in author_freq.items():
    print(f'{author}: {freq}')
```

```python
# Get the most common author names
common_authors = author_freq.most_common()  # Get only the top 3

# Separate the author names and their frequencies
authors, frequencies = zip(*common_authors)
# Plot a bar graph
plt.figure(figsize=(10, 5))
plt.bar(authors, frequencies)
plt.xlabel('Authors')
plt.ylabel('Frequency')
plt.title('Frequency of Top 3 Authors')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the least common author names
least_common_authors = author_freq.most_common()[-3:]  # Get only the bottom 3

# Separate the author names and their frequencies
authors, frequencies = zip(*least_common_authors)

# Plot a bar graph
plt.figure(figsize=(10, 5))
plt.bar(authors, frequencies)
plt.xlabel('Authors')
plt.ylabel('Frequency')
plt.title('Frequency of Bottom 3 Authors')
plt.xticks(rotation=90)
plt.show()

```

```python
# Display the bottom 3 authors
print('Bottom 3 authors:')
for author, freq in least_common_authors:
    print(f'{author}: {freq}')
```

```python
import json

# ng dn n file JSON trong Google Drive ca bn
file_path = 'Json Data-20240601/students.json'

# c d liu t file JSON
with open(file_path, 'r') as file:
    data = json.load(file)

students = data['class']['student']
```

```python
# a. Ly ra thng tin ca cc sinh vin trong lp.
print("Thng tin ca cc sinh vin trong lp:")
for student in students:
    print(student)

# b. Ly ra im (marks) ca cc sinh vin trong lp.
marks = [student['marks'] for student in students]
print("\nim ca cc sinh vin trong lp:", marks)

# c. Ly ra tn (firstname) ca cc sinh vin c im (marks) t 90 tr ln.
high_marks_students = [student['firstname'] for student in students if student['marks'] >= 90]
print("\nTn ca cc sinh vin c im (marks) t 90 tr ln:", high_marks_students)

# d. Ly ra im (marks) ca sinh vin c h (lastname) l Kad.
marks_kad = next((student['marks'] for student in students if student['lastname'] == 'Kad'), None)
print("\nim ca sinh vin c h (lastname) l Kad:", marks_kad)

# e. Ly ra tn (firstname) ca sinh vin cui cng trong danh sch.
last_student_firstname = students[-1]['firstname']
print("\nTn ca sinh vin cui cng trong danh sch:", last_student_firstname)

# f. Ly ra s lng sinh vin c im (marks) t 90 tr ln.
count_high_marks_students = sum(1 for student in students if student['marks'] >= 90)
print("\nS lng sinh vin c im (marks) t 90 tr ln:", count_high_marks_students)

# g. Ly ra im (marks) trung bnh ca cc sinh vin trong lp.
average_marks = sum(marks) / len(marks)
print("\nim (marks) trung bnh ca cc sinh vin trong lp:", average_marks)

# h. Ly ra h tn vit lin (firstname, lastname) ca sinh vin cui cng trong danh sch.
last_student_fullname = students[-1]['firstname'] + " " + students[-1]['lastname']
print("\nH tn vit lin ca sinh vin cui cng trong danh sch:", last_student_fullname)
```

```python
file_path_books = 'Json Data-20240601/books.json'


# c d liu t file JSON
with open(file_path_books, 'r') as file:
    books_data = json.load(file)

books = books_data['bookstore']['book']

# a. Ly ra tt c cc node con ca node book.
all_books = books
print("Tt c cc node con ca node book:")
for book in all_books:
    print(book)

# b. Ly ra tt c cc tn sch (title).
titles = [book['title'] for book in books]
print("\nTt c cc tn sch (title):", titles)

# c. Ly ra tn sch (title) ca quyn sch u tin.
first_title = books[0]['title']
print("\nTn sch ca quyn sch u tin:", first_title)

# d. Ly ra thng tin ca hai quyn sch u tin.
first_two_books = books[:2]
print("\nThng tin ca hai quyn sch u tin:", first_two_books)

# e. Ly ra tn sch (title) ca quyn sch cui cng.
last_title = books[-1]['title']
print("\nTn sch ca quyn sch cui cng:", last_title)

# f. Ly ra tt c cc gi tin (price).
prices = [book['price'] for book in books]
print("\nTt c cc gi tin (price):", prices)

# g. Ly ra cc gi tin (price) trn 35.
prices_over_35 = [book['price'] for book in books if book['price'] > 35]
print("\nCc gi tin (price) trn 35:", prices_over_35)

# h. Ly ra tn (title) ca cc quyn sch c gi tin (price) trn 35.
titles_over_35 = [book['title'] for book in books if book['price'] > 35]
print("\nTn ca cc quyn sch c gi tin (price) trn 35:", titles_over_35)

# i. Ly ra gi tin (price) ca cc quyn sch c gi tin (price) di 30 hoc trn 40.
prices_under_30_or_over_40 = [book['price'] for book in books if book['price'] < 30 or book['price'] > 40]
print("\nGi tin ca cc quyn sch c gi tin (price) di 30 hoc trn 40:", prices_under_30_or_over_40)

# j. Ly ra tn (title) ca cc quyn sch c gi tin (price) nm trong khong t 40 n 50.
titles_40_to_50 = [book['title'] for book in books if 40 <= book['price'] <= 50]
print("\nTn ca cc quyn sch c gi tin (price) nm trong khong t 40 n 50:", titles_40_to_50)

# k. Ly ra tn tc gi (author) ca cc quyn sch.
authors = [book['author'] for book in books]
print("\nTn tc gi (author) ca cc quyn sch:", authors)

# l. Ly ra tn (title) ca quyn sch c tn tc gi (author) l James Linn.
titles_by_james_linn = [book['title'] for book in books if 'James Linn' in (book['author'] if isinstance(book['author'], list) else [book['author']])]
print("\nTn ca quyn sch c tn tc gi (author) l James Linn:", titles_by_james_linn)

# m. Ly ra tn (title) ca quyn sch c tn tc gi (author) l J K. Rowling v gi tin (price) di 50.
titles_by_rowling_under_50 = [book['title'] for book in books if book['author'] == 'J K. Rowling' and book['price'] < 50]
print("\nTn ca quyn sch c tn tc gi (author) l J K. Rowling v gi tin (price) di 50:", titles_by_rowling_under_50)

# n. Ly ra gi bn cao nht trong s cc quyn sch.
max_price = max(prices)
print("\nGi bn cao nht trong s cc quyn sch:", max_price)

# o. Ly ra gi bn trung bnh ca cc quyn sch.
average_price = sum(prices) / len(prices)
print("\nGi bn trung bnh ca cc quyn sch:", average_price)

# p. Ly ra nm xut bn (year) ca quyn sch c tn (title) l Learning XML.
year_learning_xml = next((book['year'] for book in books if book['title'] == 'Learning XML'), None)
print("\nNm xut bn ca quyn sch c tn (title) l Learning XML:", year_learning_xml)
```
