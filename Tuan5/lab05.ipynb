```python
import requests
from bs4 import BeautifulSoup

# Send a GET request
response = requests.get('https://openaccess.thecvf.com/WACV2024')

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the <dt class="ptitle"> tags
paper_titles_tags = soup.find_all('dt', class_='ptitle')

# Extract the text from these tags
paper_titles = [tag.text.strip() for tag in paper_titles_tags]

# Print the paper titles
for title in paper_titles:
    print(title)
```

```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download the stopwords from NLTK
nltk.download('punkt')
nltk.download('stopwords')

# Define a list of unimportant words in the field of technology
unimportant_words = ['the', 'a', 'an', 'in', 'is', 'of', 'and', 'for']

# Get the list of stopwords in English
stop_words = set(stopwords.words('english'))

# Add the unimportant words to the list of stopwords
stop_words.update(unimportant_words)

# Process each title
processed_titles = []
for title in paper_titles:
    # Tokenize the title into words
    words = word_tokenize(title)
    
    # Convert each word to lowercase and remove punctuation
    words = [word.lower() for word in words if word.isalpha()]
    
    # Remove stopwords and unimportant words
    words = [word for word in words if word not in stop_words]
    
    # Add the processed title to the list
    processed_titles.append(words)

# Print the processed titles
for title in processed_titles:
    print(title)
```

```python
import collections

# Flatten the list of processed titles into a single list of keywords
keywords = [word for title in processed_titles for word in title]

# Count the frequency of each keyword
keyword_freq = collections.Counter(keywords)

# Print the frequency of each keyword
for keyword, freq in keyword_freq.items():
    print(f'{keyword}: {freq}')
```

```python
import matplotlib.pyplot as plt

# Get the most common keywords
common_keywords = keyword_freq.most_common()

# Separate the keywords and their frequencies
keywords, frequencies = zip(*common_keywords)

# Plot a bar graph
plt.figure(figsize=(10, 5))
plt.bar(keywords, frequencies)
plt.xlabel('Keywords')
plt.ylabel('Frequency')
plt.title('Frequency of Keywords')
plt.xticks(rotation=90)
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
import collections
import matplotlib.pyplot as plt

# Send a GET request
response = requests.get('https://openaccess.thecvf.com/WACV2024')

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all the <form> tags with class 'authsearch'
author_tags = soup.find_all('form', class_='authsearch')

# Extract the author names from these tags
author_names = [tag.input['value'] for tag in author_tags]

for author in author_names:
    print(author)
```

```python
# Count the frequency of each author name
author_freq = collections.Counter(author_names)

for author, freq in author_freq.items():
    print(f'{author}: {freq}')
```

```python
# Get the most common author names
common_authors = author_freq.most_common()  # Get only the top 3

# Separate the author names and their frequencies
authors, frequencies = zip(*common_authors)
# Plot a bar graph
plt.figure(figsize=(10, 5))
plt.bar(authors, frequencies)
plt.xlabel('Authors')
plt.ylabel('Frequency')
plt.title('Frequency of Top 3 Authors')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Get the least common author names
least_common_authors = author_freq.most_common()[-3:]  # Get only the bottom 3

# Separate the author names and their frequencies
authors, frequencies = zip(*least_common_authors)

# Plot a bar graph
plt.figure(figsize=(10, 5))
plt.bar(authors, frequencies)
plt.xlabel('Authors')
plt.ylabel('Frequency')
plt.title('Frequency of Bottom 3 Authors')
plt.xticks(rotation=90)
plt.show()

```

```python
# Display the bottom 3 authors
print('Bottom 3 authors:')
for author, freq in least_common_authors:
    print(f'{author}: {freq}')
```

```python
import json

# Đường dẫn đến file JSON trong Google Drive của bạn
file_path = 'Json Data-20240601/students.json'

# Đọc dữ liệu từ file JSON
with open(file_path, 'r') as file:
    data = json.load(file)

students = data['class']['student']
```

```python
# a. Lấy ra thông tin của các sinh viên trong lớp.
print("Thông tin của các sinh viên trong lớp:")
for student in students:
    print(student)

# b. Lấy ra điểm (marks) của các sinh viên trong lớp.
marks = [student['marks'] for student in students]
print("\nĐiểm của các sinh viên trong lớp:", marks)

# c. Lấy ra tên (firstname) của các sinh viên có điểm (marks) từ 90 trở lên.
high_marks_students = [student['firstname'] for student in students if student['marks'] >= 90]
print("\nTên của các sinh viên có điểm (marks) từ 90 trở lên:", high_marks_students)

# d. Lấy ra điểm (marks) của sinh viên có họ (lastname) là Kad.
marks_kad = next((student['marks'] for student in students if student['lastname'] == 'Kad'), None)
print("\nĐiểm của sinh viên có họ (lastname) là Kad:", marks_kad)

# e. Lấy ra tên (firstname) của sinh viên cuối cùng trong danh sách.
last_student_firstname = students[-1]['firstname']
print("\nTên của sinh viên cuối cùng trong danh sách:", last_student_firstname)

# f. Lấy ra số lượng sinh viên có điểm (marks) từ 90 trở lên.
count_high_marks_students = sum(1 for student in students if student['marks'] >= 90)
print("\nSố lượng sinh viên có điểm (marks) từ 90 trở lên:", count_high_marks_students)

# g. Lấy ra điểm (marks) trung bình của các sinh viên trong lớp.
average_marks = sum(marks) / len(marks)
print("\nĐiểm (marks) trung bình của các sinh viên trong lớp:", average_marks)

# h. Lấy ra họ tên viết liền (firstname, lastname) của sinh viên cuối cùng trong danh sách.
last_student_fullname = students[-1]['firstname'] + " " + students[-1]['lastname']
print("\nHọ tên viết liền của sinh viên cuối cùng trong danh sách:", last_student_fullname)
```

```python
file_path_books = 'Json Data-20240601/books.json'


# Đọc dữ liệu từ file JSON
with open(file_path_books, 'r') as file:
    books_data = json.load(file)

books = books_data['bookstore']['book']

# a. Lấy ra tất cả các node con của node book.
all_books = books
print("Tất cả các node con của node book:")
for book in all_books:
    print(book)

# b. Lấy ra tất cả các tên sách (title).
titles = [book['title'] for book in books]
print("\nTất cả các tên sách (title):", titles)

# c. Lấy ra tên sách (title) của quyển sách đầu tiên.
first_title = books[0]['title']
print("\nTên sách của quyển sách đầu tiên:", first_title)

# d. Lấy ra thông tin của hai quyển sách đầu tiên.
first_two_books = books[:2]
print("\nThông tin của hai quyển sách đầu tiên:", first_two_books)

# e. Lấy ra tên sách (title) của quyển sách cuối cùng.
last_title = books[-1]['title']
print("\nTên sách của quyển sách cuối cùng:", last_title)

# f. Lấy ra tất cả các giá tiền (price).
prices = [book['price'] for book in books]
print("\nTất cả các giá tiền (price):", prices)

# g. Lấy ra các giá tiền (price) trên 35.
prices_over_35 = [book['price'] for book in books if book['price'] > 35]
print("\nCác giá tiền (price) trên 35:", prices_over_35)

# h. Lấy ra tên (title) của các quyển sách có giá tiền (price) trên 35.
titles_over_35 = [book['title'] for book in books if book['price'] > 35]
print("\nTên của các quyển sách có giá tiền (price) trên 35:", titles_over_35)

# i. Lấy ra giá tiền (price) của các quyển sách có giá tiền (price) dưới 30 hoặc trên 40.
prices_under_30_or_over_40 = [book['price'] for book in books if book['price'] < 30 or book['price'] > 40]
print("\nGiá tiền của các quyển sách có giá tiền (price) dưới 30 hoặc trên 40:", prices_under_30_or_over_40)

# j. Lấy ra tên (title) của các quyển sách có giá tiền (price) nằm trong khoảng từ 40 đến 50.
titles_40_to_50 = [book['title'] for book in books if 40 <= book['price'] <= 50]
print("\nTên của các quyển sách có giá tiền (price) nằm trong khoảng từ 40 đến 50:", titles_40_to_50)

# k. Lấy ra tên tác giả (author) của các quyển sách.
authors = [book['author'] for book in books]
print("\nTên tác giả (author) của các quyển sách:", authors)

# l. Lấy ra tên (title) của quyển sách có tên tác giả (author) là James Linn.
titles_by_james_linn = [book['title'] for book in books if 'James Linn' in (book['author'] if isinstance(book['author'], list) else [book['author']])]
print("\nTên của quyển sách có tên tác giả (author) là James Linn:", titles_by_james_linn)

# m. Lấy ra tên (title) của quyển sách có tên tác giả (author) là J K. Rowling và giá tiền (price) dưới 50.
titles_by_rowling_under_50 = [book['title'] for book in books if book['author'] == 'J K. Rowling' and book['price'] < 50]
print("\nTên của quyển sách có tên tác giả (author) là J K. Rowling và giá tiền (price) dưới 50:", titles_by_rowling_under_50)

# n. Lấy ra giá bán cao nhất trong số các quyển sách.
max_price = max(prices)
print("\nGiá bán cao nhất trong số các quyển sách:", max_price)

# o. Lấy ra giá bán trung bình của các quyển sách.
average_price = sum(prices) / len(prices)
print("\nGiá bán trung bình của các quyển sách:", average_price)

# p. Lấy ra năm xuất bản (year) của quyển sách có tên (title) là Learning XML.
year_learning_xml = next((book['year'] for book in books if book['title'] == 'Learning XML'), None)
print("\nNăm xuất bản của quyển sách có tên (title) là Learning XML:", year_learning_xml)
```
